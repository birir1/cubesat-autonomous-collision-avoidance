```python
## Nessesary Import Libraries

import os
import requests
from tqdm import tqdm
import pandas as pd
```

```python
## Define Projects Path

BASE_PATH = "/root/workspace/cubesat_autonomous_collision_avoidance"

RAW_DATA_PATH = os.path.join(BASE_PATH, "data", "raw")

TLE_PATH = os.path.join(RAW_DATA_PATH, "tle")
CONJUNCTION_PATH = os.path.join(RAW_DATA_PATH, "conjunction")
DEBRIS_PATH = os.path.join(RAW_DATA_PATH, "debris")

os.makedirs(TLE_PATH, exist_ok=True)
os.makedirs(CONJUNCTION_PATH, exist_ok=True)
os.makedirs(DEBRIS_PATH, exist_ok=True)

print("Directories created successfully")
```

```python
## Helper Download Function

def download_file(url, save_path):

    response = requests.get(url, stream=True)

    total_size = int(response.headers.get('content-length', 0))

    with open(save_path, "wb") as file, tqdm(
        desc=save_path,
        total=total_size,
        unit="iB",
        unit_scale=True,
        unit_divisor=1024,
    ) as bar:

        for data in response.iter_content(chunk_size=1024):
            size = file.write(data)
            bar.update(size)

    print(f"Downloaded: {save_path}")
```

```python
## Download TLE Orbital Dataset

tle_urls = {
    "active_satellites": "https://celestrak.org/NORAD/elements/gp.php?GROUP=active&FORMAT=tle",
    "starlink": "https://celestrak.org/NORAD/elements/gp.php?GROUP=starlink&FORMAT=tle",
    "oneweb": "https://celestrak.org/NORAD/elements/gp.php?GROUP=oneweb&FORMAT=tle"
}

for name, url in tle_urls.items():

    save_path = os.path.join(TLE_PATH, f"{name}.txt")

    download_file(url, save_path)
```

```python
## Download Space Debris Dataset

debris_urls = {
    "visual_satellites": "https://celestrak.org/NORAD/elements/gp.php?GROUP=visual&FORMAT=tle",
    "iridium_debris": "https://celestrak.org/NORAD/elements/gp.php?GROUP=iridium-33-debris&FORMAT=tle"
}

for name, url in debris_urls.items():

    save_path = os.path.join(DEBRIS_PATH, f"{name}.txt")

    download_file(url, save_path)
```

```python
## Download Satellite Metadata Dataset

ucs_url = "https://www.ucsusa.org/sites/default/files/2023-09/UCS-Satellite-Database.csv"

save_path = os.path.join(RAW_DATA_PATH, "satellite_metadata.csv")

download_file(ucs_url, save_path)
```
